```python
import subprocess
import os
import shutil
```

```python
N = 1    # number of replicates

# run simulations
for n in range(N):
    for s in range(1, 11):
        command = f"python adhesion/abm.py -n adhesion_{s}_{n} -m 0 -s {s}"
        subprocess.run(command.split(" "), stdout=subprocess.DEVNULL)    # ignores output
        
    print(f"\rProgress: {n+1} of {N}", end="")
```

```python
directory = "outputs"

# consolidate CSVs generated by simulations
sims = os.listdir(directory)
for name in sims:
    if name.startswith("adhesion"):
        # get path to agent data in last step of simulation and copy
        path = f"{directory}/{name}/{name}_values/{name}_values_500.csv"
        shutil.copy(path, f"csvs/{name}.csv")
```
